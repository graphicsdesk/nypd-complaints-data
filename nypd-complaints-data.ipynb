The second section covers complaints and arrests made in the 4 weeks before and after a major event (an event that sparked much unrest/protest) regarding police brutality. 

#### Events mentioned:
- The start of national coverage (and protest) of the Trayvon Martin Shooting
- The acquittal of George Zimmerman
- The Grand Jury's decision to not indict Michael Brown's shooter
- The death of Freddie Gray

```python
import pandas as pd
weekList=['week1','week2','week3','week4','week5','week6','week7','week8']
yearList=['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','Total/Average']
```

```python
arrcd9=[]
barrcd9=[]
compcd9=[]
bcompcd9=[]
arrM=[]
barrM=[]
compM=[]
bcompM=[]
pbarrcd9=[]
pbcompcd9=[]
pbarrm=[]
pbcompm=[]
for x in range(len(weekList)):
    #CD9
    arrDF = pd.read_json(r'JSONs/Arrests/Martin2012/cd9/'+weekList[x]+'.json')
    arrDF = arrDF[arrDF.PERP_RACE!='UNKNOWN']
    arrDF= arrDF.dropna()
    arrcd9.append(len(arrDF))
    arrDF=arrDF[arrDF.PERP_RACE=='BLACK']
    barrcd9.append(len(arrDF))
    compDF = pd.read_json(r'JSONs/Complaints/Martin2012/cd9/'+weekList[x]+'.json')
    compDF = compDF[compDF.SUSP_RACE!='UNKNOWN']
    compDF = compDF.dropna()
    compcd9.append(len(compDF))
    compDF=compDF[compDF.SUSP_RACE=='BLACK']
    bcompcd9.append(len(compDF))
    pbarrcd9.append(barrcd9[len(barrcd9)-1]/arrcd9[len(arrcd9)-1]*100)
    pbcompcd9.append(bcompcd9[len(bcompcd9)-1]/compcd9[len(compcd9)-1]*100)
    #Manhattan
    arrDF = pd.read_json(r'JSONs/Arrests/Martin2012/manhattan/'+weekList[x]+'.json')
    arrDF = arrDF[arrDF.PERP_RACE!='UNKNOWN']
    arrDF= arrDF.dropna()
    arrM.append(len(arrDF))
    arrDF=arrDF[arrDF.PERP_RACE=='BLACK']
    barrM.append(len(arrDF))
    compDF = pd.read_json(r'JSONs/Complaints/Martin2012/manhattan/'+weekList[x]+'.json')
    compDF = compDF[compDF.SUSP_RACE!='UNKNOWN']
    compDF = compDF.dropna()
    compM.append(len(compDF))
    compDF=compDF[compDF.SUSP_RACE=='BLACK']
    bcompM.append(len(compDF))
    pbarrm.append(barrM[len(barrcd9)-1]/arrM[len(arrcd9)-1]*100)
    pbcompm.append(bcompM[len(bcompcd9)-1]/compM[len(compcd9)-1]*100)
d={'Complaints w/ Black Suspects':bcompcd9,'Total Complaints':compcd9,'Percentage of Black-Suspect Complaints':pbcompcd9,'Arrests w/ Black Perpetrators':barrcd9,'Total Arrests':arrcd9,'Percentage of Black-Perpetrator Arrests':pbarrcd9}
e={'Complaints w/ Black Suspects':bcompM,'Total Complaints':compM,'Percentage of Black-Suspect Complaints':pbcompm,'Arrests w/ Black Perpetrators':barrM,'Total Arrests':arrM,'Percentage of Black-Perpetrator Arrests':pbarrm}
martinDFcd9=pd.DataFrame(data=d)
martinDFm=pd.DataFrame(data=e)
```

```python
compcd9=[]
bcompcd9=[]
wcompcd9=[]
pbcompcd9=[]
pwcompcd9=[]
totcd9=0
totbcd9=0
totwcd9=0
for x in range(len(yearList)-1):
    #Black
    arrDF = pd.read_json(r'JSONs/Complaints/Yearly/cd9/'+yearList[x]+'.json')
    arrDF = arrDF[arrDF.SUSP_RACE!='UNKNOWN']
    arrDF= arrDF.dropna()
    compcd9.append(len(arrDF))
    totcd9+=len(arrDF)
    arrDF=arrDF[arrDF.SUSP_RACE=='BLACK']
    bcompcd9.append(len(arrDF))
    totbcd9+=len(arrDF)
    pbcompcd9.append(bcompcd9[len(bcompcd9)-1]/compcd9[len(compcd9)-1]*100)
    arrDF = pd.read_json(r'JSONs/Complaints/Yearly/cd9/'+yearList[x]+'.json')
    arrDF = arrDF[arrDF.SUSP_RACE!='UNKNOWN']
    arrDF= arrDF.dropna()
    arrDF=arrDF[arrDF.SUSP_RACE=='WHITE']
    wcompcd9.append(len(arrDF))
    totwcd9+=len(arrDF)
    pwcompcd9.append(wcompcd9[len(bcompcd9)-1]/compcd9[len(compcd9)-1]*100)
compcd9.append(totcd9)
bcompcd9.append(totbcd9)
wcompcd9.append(totwcd9)
pbcompcd9.append(totbcd9/totcd9*100)
pwcompcd9.append(totwcd9/totcd9*100)

d={'Total Complaints':compcd9,'Complaints w/ Black Suspects':bcompcd9,'Percentage of Black-Suspect Complaints':pbcompcd9,'Complaints w/ White Suspects':wcompcd9,'Percentage of White-Suspect Complaints':pwcompcd9}
cd9Years=pd.DataFrame(data=d,index=yearList)
```
